```python
import pandas as pd

skip_matches = [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]
delivery_data = []

for _, row in matches.iterrows():
    match_id = row['match_id']
    if match_id in skip_matches:
        continue
    
    try:
        deliveries = row['innings'][0]['1st innings']['deliveries']
        batting_team = row['innings'][0]['1st innings']['team']
    except (KeyError, IndexError, TypeError):
        continue  # Skip malformed records

    for delivery in deliveries:
        for ball, detail in delivery.items():
            delivery_data.append({
                'match_id': match_id,
                'teams': row.get('info.teams', []),
                'batting_team': batting_team,
                'ball': ball,
                'batsman': detail.get('batsman'),
                'bowler': detail.get('bowler'),
                'runs': detail.get('runs', {}).get('total', 0),
                'player_dismissed': detail.get('wicket', {}).get('player_out', None),
                'city': row.get('info.city'),
                'venue': row.get('info.venue')
            })

delivery_df = pd.DataFrame(delivery_data)
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```
